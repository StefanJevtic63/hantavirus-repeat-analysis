# Initial stage - dataset creation and statistics

```python
import os
import pandas as pd

CURRENT_DIR = os.path.dirname(os.path.abspath("__file__"))
DATASET_DIR = os.path.dirname(CURRENT_DIR)
DATASET_PATH = os.path.join(DATASET_DIR, 'data', "dataset_all_sequences.csv")

RANDOM_STATE = 324
```

```python
df: pd.DataFrame = pd.read_csv(DATASET_PATH)
df.info()
```

```python
df.describe()
```

```python
df.head()
```

# Preprocessing 

We will remove the 'Sequence_ID' column since it provides no value to the models

```python
df.drop(columns=['Sequence_ID'], inplace=True)
```

```python
feature_names = df.columns[:-1]
feature_names
```

Filter out rows where all repeat characteristics sum up to zero (empty sequences)

```python
df = df[df[feature_names].sum(axis=1) > 0]
```

```python
df["Virus_Species"].value_counts()
```

```python
df["Virus_Species"] = df["Virus_Species"].replace("orthohantavirus hantanense", "Orthohantavirus hantanense")
df["Virus_Species"].value_counts()
```

We will only keep the following classes: <br>
Orthohantavirus puumalaense, Orthohantavirus dobravaense, Orthohantavirus hantanense and Orthohantavirus sinnombreense

```python
valid_classes = [
    "Orthohantavirus puumalaense", 
    "Orthohantavirus dobravaense", 
    "Orthohantavirus hantanense",
    "Orthohantavirus sinnombreense"
]
df.drop(df[~df["Virus_Species"].isin(valid_classes)].index, axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df["Virus_Species"].value_counts()
```

```python
X = df[feature_names]
y_raw = df["Virus_Species"]
```

Encode categorical text labels into unique numerical integers


```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y_raw)
```

```python
from sklearn.model_selection import train_test_split

X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.1, random_state=RANDOM_STATE, stratify=y
)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.1, random_state=RANDOM_STATE, stratify=y_temp
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

print("Training features matrix shape:", X_train_scaled.shape)
print("Validation features matrix shape:", X_val_scaled.shape)
print("Testing features matrix shape:", X_test_scaled.shape)
print("Training labels shape:", y_train.shape)
print("Validation labels shape:", y_val.shape)
print("Testing labels shape:", y_test.shape)
```

We will apply SMOTE (Synthetic Minority Over-sampling Technique) to perform over sampling to fight the class imbalance 

```python
from imblearn.over_sampling import SMOTE
import numpy as np

smote = SMOTE(random_state=RANDOM_STATE)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_scaled, y_train)

print("Balanced training features matrix shape:", X_train_resampled.shape)
print("Balanced training labels shape:", y_train_resampled.shape)
```

```python
unique, counts = np.unique(y_train_resampled, return_counts=True)
for label_idx, count in zip(unique, counts):
    species_name = label_encoder.inverse_transform([label_idx])[0]
    print(f"{species_name}: {count}")
```

# Creating the classification models

```python
import numpy as np
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

X_train_resampled = pd.DataFrame(X_train_resampled, columns=feature_names)
X_val_scaled = pd.DataFrame(X_val_scaled, columns=feature_names)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=feature_names)

tuning_grid = {
    "Random Forest": {"max_depth": [10, 20, None], "min_samples_split": [2, 5]},
    "XGBoost": {"max_depth": [3, 6, 9], "learning_rate": [0.05, 0.1]},
    "LightGBM": {"num_leaves": [15, 30, 60], "learning_rate": [0.05, 0.1]},
    "K-Nearest Neighbors (KNN)": {"n_neighbors": [3, 5, 11], "weights": ["uniform", "distance"]},
    "Linear SVM": {"C": [0.1, 1.0, 10.0]}
}

optimal_models = {}
accuracy_results = {}
confusion_matrices = {}
```

### KNN

```python
best_knn_acc = -1
best_knn_params = {}

for k in tuning_grid["K-Nearest Neighbors (KNN)"]["n_neighbors"]:
    for w in tuning_grid["K-Nearest Neighbors (KNN)"]["weights"]:
        clf = KNeighborsClassifier(n_neighbors=k, weights=w, n_jobs=-1)
        clf.fit(X_train_resampled, y_train_resampled)
        val_acc = accuracy_score(y_val, clf.predict(X_val_scaled))
        if val_acc > best_knn_acc:
            best_knn_acc = val_acc
            best_knn_params = {"n_neighbors": k, "weights": w}

print(f"Optimal KNN Parameters: {best_knn_params}")
knn_final = KNeighborsClassifier(**best_knn_params, n_jobs=-1)
knn_final.fit(X_train_resampled, y_train_resampled)
optimal_models["K-Nearest Neighbors (KNN)"] = knn_final
```

### Random Forest

```python
best_rf_acc = -1
best_rf_params = {}

for depth in tuning_grid["Random Forest"]["max_depth"]:
    for split in tuning_grid["Random Forest"]["min_samples_split"]:
        clf = RandomForestClassifier(max_depth=depth, min_samples_split=split, random_state=42, n_jobs=-1)
        clf.fit(X_train_resampled, y_train_resampled)
        val_acc = accuracy_score(y_val, clf.predict(X_val_scaled))
        if val_acc > best_rf_acc:
            best_rf_acc = val_acc
            best_rf_params = {"max_depth": depth, "min_samples_split": split}

print(f"Optimal Random Forest Parameters: {best_rf_params}")
rf_final = RandomForestClassifier(**best_rf_params, random_state=42, n_jobs=-1)
rf_final.fit(X_train_resampled, y_train_resampled)
optimal_models["Random Forest"] = rf_final
```

### XGBoost

```python
EARLY_STOPPING_ROUNDS = 10
N_ESTIMATORS = 500

best_xgb_acc = -1
best_xgb_params = {}

for depth in tuning_grid["XGBoost"]["max_depth"]:
    for lr in tuning_grid["XGBoost"]["learning_rate"]:
        clf = XGBClassifier(
            max_depth=depth, 
            learning_rate=lr, 
            n_estimators=N_ESTIMATORS, 
            random_state=RANDOM_STATE, 
            eval_metric="mlogloss", 
            early_stopping_rounds=EARLY_STOPPING_ROUNDS, 
            n_jobs=-1
        )
        clf.fit(X_train_resampled, y_train_resampled, eval_set=[(X_val_scaled, y_val)], verbose=False)
        
        val_acc = accuracy_score(y_val, clf.predict(X_val_scaled))
        if val_acc > best_xgb_acc:
            best_xgb_acc = val_acc
            best_xgb_params = {"max_depth": depth, "learning_rate": lr, "n_estimators": clf.best_iteration + 1}

print(f"Optimal XGBoost Parameters: {best_xgb_params}")

xgb_final = XGBClassifier(
    max_depth=best_xgb_params["max_depth"], 
    learning_rate=best_xgb_params["learning_rate"], 
    n_estimators=best_xgb_params["n_estimators"], 
    random_state=RANDOM_STATE, 
    eval_metric="mlogloss", 
    n_jobs=-1
)
xgb_final.fit(X_train_resampled, y_train_resampled)
optimal_models["XGBoost"] = xgb_final
```

### LightBGM

```python
N_ESTIMATORS = 500

best_lgb_acc = -1
best_lgb_params = {}

for leaves in tuning_grid["LightGBM"]["num_leaves"]:
    for lr in tuning_grid["LightGBM"]["learning_rate"]:
        clf = LGBMClassifier(
            num_leaves=leaves, 
            learning_rate=lr, 
            n_estimators=N_ESTIMATORS, 
            random_state=RANDOM_STATE, 
            verbosity=-1, 
            n_jobs=-1
        )
        clf.fit(X_train_resampled, y_train_resampled)
        
        val_acc = accuracy_score(y_val, clf.predict(X_val_scaled))
        if val_acc > best_lgb_acc:
            best_lgb_acc = val_acc
            best_lgb_params = {"num_leaves": leaves, "learning_rate": lr}

print(f"Optimal LightGBM Parameters: {best_lgb_params}")
lgb_final = LGBMClassifier(**best_lgb_params, random_state=RANDOM_STATE, verbosity=-1, n_jobs=-1)
lgb_final.fit(X_train_resampled, y_train_resampled)
optimal_models["LightGBM"] = lgb_final
```

### Linear SVM

```python
MAX_ITER = 10000

best_svm_acc = -1
best_svm_params = {}

for c_val in tuning_grid["Linear SVM"]["C"]:
    clf = LinearSVC(C=c_val, random_state=RANDOM_STATE, max_iter=MAX_ITER)
    clf.fit(X_train_resampled, y_train_resampled)
    val_acc = accuracy_score(y_val, clf.predict(X_val_scaled))
    if val_acc > best_svm_acc:
         best_svm_acc = val_acc
         best_svm_params = {"C": c_val}

print(f"Optimal Linear SVM Parameters: {best_svm_params}")
svm_final = LinearSVC(**best_svm_params, random_state=RANDOM_STATE, max_iter=MAX_ITER)
svm_final.fit(X_train_resampled, y_train_resampled)
optimal_models["Linear SVM"] = svm_final
```

# Evaluation

```python
def evaluate_model(model, name, X_test, y_test, label_encoder):
    """
    Evaluates a trained model on the test dataset, computes performance metrics,
    stores the results globally, and prints a comprehensive classification report.

    Parameters:
        model : estimator object
            The trained classification model to evaluate.
        name : str
            The identifier name of the model used as a key for storing results.
        X_test : array-like or DataFrame
            The scaled feature matrix for the testing dataset.
        y_test : array-like
            The true integer labels for the testing dataset.
        label_encoder : LabelEncoder
            The fitted LabelEncoder instance used to map integer labels back 
            to the original biological virus species names.

    Returns:
        None
    """
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred, target_names=label_encoder.classes_)
    
    accuracy_results[name] = acc
    confusion_matrices[name] = conf_matrix

    print(f"Final Evaluation for optimized model: {name}")
    print(class_report)
```

```python
name = "K-Nearest Neighbors (KNN)"
model = optimal_models[name]
evaluate_model(model, name, X_test_scaled, y_test, label_encoder)
```

# Plot the accuracy performance

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 5))
sns.barplot(x=list(accuracy_results.keys()), y=list(accuracy_results.values()), palette="YlGnBu")
plt.title("Test Accuracy for Models Optimized via Validation Splits", fontsize=14, pad=15)
plt.ylabel("Accuracy Score (0.0 - 1.0)", fontsize=12)
plt.xlabel("Optimized Classifiers", fontsize=12)
plt.ylim(0, 1.1)

for idx, val in enumerate(accuracy_results.values()):
    plt.text(idx, val + 0.02, f"{val:.4f}", ha="center", fontweight="bold", fontsize=11)

plt.tight_layout()
plt.show()
```

```python
def plot_single_confusion_matrix(matrix, model_name, class_names):
    """
    Plots a single confusion matrix heatmap using a standardized professional palette.

    Args:
        matrix: The calculated confusion matrix array.
        model_name: String name of the model for the plot title.
        class_names: List of target class labels for axis ticks.
    """
    plt.figure(figsize=(7, 5))
    sns.heatmap(
        matrix, 
        annot=True, 
        fmt="d", 
        cmap="YlGnBu", 
        xticklabels=class_names, 
        yticklabels=class_names,
        cbar=False,
        annot_kws={"size": 12, "weight": "bold"}
    )
    plt.title(f"Confusion Matrix: {model_name}", fontsize=13, pad=12, fontweight="bold")
    plt.ylabel("Ground Truth Class", fontsize=11)
    plt.xlabel("Predicted Class", fontsize=11)
    plt.xticks(rotation=30, ha="right")
    plt.tight_layout()
    plt.show()
```

```python
plot_single_confusion_matrix(
    confusion_matrices["K-Nearest Neighbors (KNN)"], 
    "K-Nearest Neighbors (KNN)", 
    label_encoder.classes_
)
```
